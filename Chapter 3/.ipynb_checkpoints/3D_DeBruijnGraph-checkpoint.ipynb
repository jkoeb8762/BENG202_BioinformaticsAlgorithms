```python
def DeBruijnGraph(k, text):
    kmers = []
    n = len(text)
    #generate kmers
    for i in range(n-k+2):
        kmers.append(text[i:i+k-1])
    graphMap = {}
    for j in range(n-k+1):
        working = kmers[j]
        new = kmers[j+1]
        if working not in graphMap:
            graphMap[working] = [new]
        else:
            graphMap[working].append(new)
    return graphMap
#print(DeBruijn(4, "AAGATTCTCTAAGA"))
```
