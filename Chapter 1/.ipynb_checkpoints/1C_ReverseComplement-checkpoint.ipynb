```python
#Find the Reverse Complement of a String
def ReverseComplement(pattern):
    revString = pattern[::-1]
    revComp= ""
    for char in revString:
        if char == "A":
            revComp += "T"
        elif char == "T":
            revComp += "A"
        elif char == "C":
            revComp += "G"
        elif char == "G":
            revComp += "C"
        
    return revComp
```
