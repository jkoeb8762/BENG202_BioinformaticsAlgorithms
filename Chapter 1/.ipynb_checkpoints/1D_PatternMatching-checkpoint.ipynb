```python
#Find All Occurrences of a Pattern in a String
def PatternMatching(pattern, genome):
    index = []
    for i in range(len(genome)-len(pattern)+1):
        if genome[i:i+len(pattern)] == pattern:
            index.append(i)
    print(*index)
```

```python
g1 = ""
p1 = "CTTGATCAT"
```

```python
PatternMatching(p1, g1)
```
